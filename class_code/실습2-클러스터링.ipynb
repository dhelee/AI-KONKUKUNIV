```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
```

```python
def plotResult(X, y, title):
    # 인덱스에 조건식 쓸 수 있음!!
    # y=0인 데이터들
    plt.scatter(X[y==0,0], X[y==0,1], c='lightblue', marker='o', s=40, label='Cluster1')
    # y=1인 데이터들
    plt.scatter(X[y==1,0], X[y==1,1], c='red', marker='s', s=40, label='Cluster2')
    
    plt.title(title)
    
    plt.legend()
    plt.show()
```

```python
from sklearn.cluster import KMeans

X, y = make_moons(n_samples=200, noise=0.05, random_state=0)
km = KMeans(n_clusters=2, random_state=0)
y_km = km.fit_predict(X)

plotResult(X, y_km, title='K-means Clustering Result')
```

```python
from sklearn.cluster import DBSCAN

X, _ = make_moons(n_samples=200, noise=0.05, random_state=0)
db = DBSCAN(eps=0.2, min_samples=5, metric='euclidean')
y_db = db.fit_predict(X)

plotResult(X, y_db, title='DBSCAN Clustering Result')
```
